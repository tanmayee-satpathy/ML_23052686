```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

X = np.array([
    [2,6], [3,4], [3,8], [4,7],
    [6,2], [6,4], [7,3], [7,4], [8,5],
    [20,20] 
])

db = DBSCAN(eps=2, min_samples=2)

labels = db.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels)

plt.title("DBSCAN Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

print("Cluster Labels:", labels)
```
