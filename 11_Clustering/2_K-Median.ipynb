```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([
    [2,6], [3,4], [3,8],
    [4,7], [6,2], [6,4],
    [7,3], [7,4], [8,5]
])

k = 2

medians = X[:k]

for _ in range(10):

    clusters = [[] for _ in range(k)]

    for point in X:
        distances = [np.sum(np.abs(point - m)) for m in medians]
        idx = np.argmin(distances)
        clusters[idx].append(point)

    new_medians = []
    for cluster in clusters:
        cluster = np.array(cluster)
        new_medians.append(np.median(cluster, axis=0))

    medians = np.array(new_medians)

clusters = [np.array(c) for c in clusters]

for cluster in clusters:
    plt.scatter(cluster[:,0], cluster[:,1])

plt.scatter(medians[:,0], medians[:,1], marker='X', s=200)

plt.title("K-Median Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
